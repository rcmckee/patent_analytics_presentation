# intro to webscraping

I am using [mechanicalsoup](https://mechanicalsoup.readthedocs.io/en/stable/tutorial.html#first-contact-step-by-step) for this example, but websites with javascript, or to automate form fills online, checkout [selenium](https://selenium-python.readthedocs.io/). Other tools include [beautifulsoup](https://www.crummy.com/software/BeautifulSoup/bs4/doc/), and [crawlera](https://www.scrapinghub.com/crawlera/). Finally, use [regex101.com](https://www.regex101.com) to test regex options.

```
!pip install mechanicalsoup
import mechanicalsoup
import re
```

websitescreenshot.png

```
#What a website looks like when we automate it
browser = mechanicalsoup.StatefulBrowser(user_agent="MechanicalSoup")
browser.open("https://www.robertcmckee.com")
browser.get_current_page()
```

## Get all of the links from the page

```
list_of_links_on_page = browser.links()
list_of_links_on_page
```

## Get text from the paragraphs on the page

```
for p in browser.get_current_page().find_all('p'):
  print(p.text)
```
